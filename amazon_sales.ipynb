```python
import psycopg2
from psycopg2 import sql
```

```python
conn = psycopg2.connect(
            dbname="postgres",        
            user="postgres",         
            password="postgres",     
            host="localhost",            
            port="5433"                   
        )

        # Create a cursor object
cur = conn.cursor()
```

```python

# Define connection parameters
def readQuery(_query):


        # Define your SQL query
    query = sql.SQL(_query) 

    # Execute the query
    cur.execute(query)

        # Fetch and print the result
    rows = cur.fetchall()
    return rows

        # Close the cursor and connection
```

```python
readQuery("select * from raw_amazon_table")
```

```python
readQuery('select * from raw_amazon_table where "Region" = \'Europe\'')
```

```python
def createQuery(_query):
    
    cur.execute(_query)
    print("Table created successfully, Great Work!")
    conn.commit()
```

```python
create_Query = """
CREATE TABLE IF NOT EXISTS public.raw_amazon_table_test1
(
    "Region" "text",
    "Country" "text",
    "ItemType" "text",
    "SalesChannel" "text",
    "OrderPriority" "text",
    "OrderDate" "text",
    "OrderId" "text",
    "ShipDate" "text",
    "UnitSold" "text",
    "UnitPrice" "text",
    "UnitCost" "text",
    "TotalRevenue" "text",
    "TotalCost" "text",
    "TotalProfit" "text"
)
"""
```

```python
createQuery(create_Query)
```

```python
def insertQuery(_query,insert_query_data):   
    
    for i in insert_query_data:
        cur.execute(_query, i)
 
        conn.commit()
        count = cur.rowcount
    print(count, "Record inserted successfully \
    into the table")
```

Which Region and Country have the highest Profit 

```python
_query = """SELECT 
    "Region", 
    "Country",
    SUM(CAST("TotalProfit" AS numeric)) AS total_profit
FROM 
    public.raw_amazon_table
GROUP BY 
    "Region", 
    "Country"
ORDER BY 
    total_profit DESC
LIMIT 5;
"""
```

```python
readQuery(_query)
```

```python
import pandas as pd

df = pd.DataFrame(readQuery(_query), columns=['Region', 'Country', 'TotalProfit'])
df
```

```python
_query = """
SELECT 
    "Region", 
    "Country",
    "OrderPriority",
    "OrderDate",
    "ShipDate",
	TO_DATE("ShipDate", 'MM/DD/YYYY') - TO_DATE("OrderDate", 'MM/DD/YYYY') AS TotalDays
FROM 
    public.raw_amazon_table
WHERE 
    "OrderPriority" = 'H';
"""
```
